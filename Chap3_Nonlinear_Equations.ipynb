```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy.linalg as LA
```

```python
def bisection(func, a, b, nmax, e):
  fa = func(a)
  fb = func(b)
  df = []
  if np.sign(fa)==np.sign(fb):
    return('function has same signs at a and b')

  for i in range(nmax):
    error = (b - a)/2
    c = a + error
    fc = func(c)
    df.append([c, fc, error, a, b])

    if np.abs(error) < e:
      df = pd.DataFrame(df, columns=["c", "f(c)", "error", "a", "b"])
      print('convergence')
      return df

    if fc == 0:
      print('{} is root of this fucntion'.format(c))
      df = pd.DataFrame(df, columns=["c", "f(c)", "error", "a", "b"])
      return df

    if np.sign(fa)==np.sign(fc):
      a = c
      fa = fc
    else:
      b = c
      fb = fc
  df = pd.DataFrame(df, columns=["c", "f(c)", "error", "a", "b"])
  return df
```

```python
def false_position(func, a, b, nmax, e):
  fa = func(a)
  fb = func(b)
  df = []
  if np.sign(fa)==np.sign(fb):
    return('function has same signs at a and b')

  for i in range(nmax):
    if np.sign(fa)==np.sign(fb):
      c = (a*fb - 2*b*fa)/(fb-2*fa)
    else:
      c = (2*a*fb - b*fa)/(2*fb-fa)

    error = (b - a)/2
    fc = func(c)
    df.append([c, fc, error, a, b])

    if np.abs(error) < e:
      df = pd.DataFrame(df, columns=["c", "f(c)", "error", "a", "b"])
      print('convergence')
      return df

    if fc == 0:
      print('{} is root of this fucntion'.format(c))
      df = pd.DataFrame(df, columns=["c", "f(c)", "error", "a", "b"])
      return df

    if np.sign(fa)==np.sign(fc):
      a = c
      fa = fc
    else:
      b = c
      fb = fc
  df = pd.DataFrame(df, columns=["c", "f(c)", "error", "a", "b"])
  return df
```

```python
def test_fbm(range, n, e):
  fbm1 = false_position(f7, range[0][0], range[0][1], n[0], e[0])
  fbm2 = false_position(g7, range[1][0], range[1][1], n[1], e[1])
  fbm3 = false_position(h7, range[2][0], range[2][1], n[2], e[2])
  return fbm1.iloc[-3:],fbm2.iloc[-3:], fbm3.iloc[-3:]

test_fbm(range_list, n_list, e_list)
```

```python
def Newton(func, dfunc, x, nmax, e, delta):
  fx = func(x)
  df =[]
  df.append([x, fx, 0, 0])

  for i in range(nmax):
    fp = dfunc(x)
    if np.abs(fp) < delta:
      df = pd.DataFrame(df, columns=["x", "f(x)", '|h|', 'df'])
      print("small derivation")
      return df

    d = fx/fp
    x = x - d
    fx = func(x)
    df.append([x, fx, d, fp])

    if np.abs(d) < e:
      df = pd.DataFrame(df, columns=["x", "f(x)", '|h|', 'df'])
      print("convergence")
      return df

  df = pd.DataFrame(df, columns=["x", "f(x)", '|h|', 'df'])
  return df
```

```python
#7
def fast_Newton(func, dfunc, x, nmax, e, delta):
  fx = func(x)
  df =[]
  df.append([x, fx, 0, 0])

  for i in range(nmax):
    fp = dfunc(x)
    if np.abs(fp) < delta:
      df = pd.DataFrame(df, columns=["x", "f(x)", '|h|', 'df'])
      print("small derivation")
      return df

    d = fx/fp
    if np.abs(func(x-d)) < np.abs(fx):
      x = x - d
    else:
      x = x - 0.5*d

    fx = func(x)

    df.append([x, fx, d, fp])

    if np.abs(d) < e:
      df = pd.DataFrame(df, columns=["x", "f(x)", '|h|', 'df'])
      print("convergence")
      return df

  df = pd.DataFrame(df, columns=["x", "f(x)", '|h|', 'df'])
  return df
```

```python
#21
def f21(x, y):
  f = x**2 + y**2 -25
  return f

def g21(x, y):
  g = x**2 - y - 2
  return g

def F21(X):
  f_1 = f21(X[0], X[1])
  f_2 = g21(X[0], X[1])
  F = np.array([f_1, f_2], dtype='float').reshape([2,1])
  return F

def dF21(X):
  F = np.array([[2*X[0], 2*X[1]],[2*X[0], -1]], dtype='float')
  return F

def simul_Newton(X, F, J, nmax):
  df = []

  for i in range(nmax):
    Fx = F(X)
    df.append([X, Fx])
    Jx = J(X)
    J_inv = LA.inv(Jx)
    H = np.dot(J_inv, Fx)
    X = X - H
  Fx = F(X)
  df.append([X, Fx])
  df = pd.DataFrame(df)
  return df

X0 = np.array([0.1, 0.1]).reshape([2, 1])
simul_Newton(X0, F21, dF21, 10)
```

```python
class real_ftn():
  def __init__(self):
    self.x =0
    self.y =0
    self.num =0
    self.dx =0
    self.dy =0

  def num(self, f):
    return self.num

  def dx(self, f):
    return self.dx

  def dy(self, f):
    return self.dy

  def setting(self, x, y, f, dx, dy):
    self.x = x
    self.y = y
    self.num = f(self.x, self.y)
    self.dx = dx(self.x, self.y)
    self.dy = dy(self.x, self.y)
```

```python
#0
def Secant(f, a, b, nmax, e):
  fa = f(a)
  fb = f(b)
  df = []
  if np.abs(fa) > np.abs(fb):
    c = a
    a = b
    b = c
    c = fa
    fa = fb
    fb = c

  df.append([a, fa])
  df.append([b, fb])

  for i in np.arange(2, nmax+1):
    if np.abs(fa) > np.abs(fb):
      c = a
      a = b
      b = c
      c = fa
      fa = fb
      fb = c

    d = (b-a)/(fb - fa)
    b = a
    fb = fa
    d = d*fa
    if np.abs(d) < e:
      #print("convergence")
      df = pd.DataFrame(df, columns=['x', 'f(x)'])
      return df

    a = a-d
    fa = f(a)
    df.append([a, fa])

  df = pd.DataFrame(df, columns=['x', 'f(x)'])
  return df
```

```python
#11
def f11(x):
  f11 = (x-10)**2
  return f11

def multi_Secant(f, a, b, nmax, e):
  fa = f(a)
  fb = f(b)
  df = []
  if np.abs(fa) > np.abs(fb):
    c = a
    a = b
    b = c
    c = fa
    fa = fb
    fb = c

  df.append([a, fa])
  df.append([b, fb])

  for i in np.arange(2, nmax+1):
    if np.abs(fa) > np.abs(fb):
      c = a
      a = b
      b = c
      c = fa
      fa = fb
      fb = c

    fc = f((a+b)/2)
    fprime_1 = (fb - fa)/(b-a)
    fprime_2 = (2*fc - fa - fb)/(0.5*(b-a)**2)
    b = a
    fb = fa
    d = fprime_1**2
    d = d - fa*fprime_2
    d = np.sqrt(d)
    d = fa / d
    if np.abs(d) < e:
      print("convergence")
      df = pd.DataFrame(df, columns=['x', 'f(x)'])
      return df

    a = a-d
    fa = f(a)
    df.append([a, fa])

  df = pd.DataFrame(df, columns=['x', 'f(x)'])
  return df
```

$f^{\prime\prime}(x_n)$ 은 아래와 같이 근사식을 사용

$$f^{\prime}(x_{n-1}) = \frac{f(x_{n-1} + h/2) - f(x_{n-1})}{h/2}$$

$$f^{\prime}(x_{n}) = \frac{f(x_{n}) - f(x_{n} - h/2)}{h/2}$$


$$f^{\prime\prime}(x_n) = \frac{f^{\prime}(x_{n-1}) - f^{\prime}(x_{n})}{x_{n-1} - x_{n}} = \frac{1}{x_{n-1} - x_{n}}{(\frac{2f(x_{n} - h/2) - f(x_{n-1})- f(x_{n})}{h/2})}$$

```python
#13
def Halley_Secant(f, a, b, nmax, e):
  fa = f(a)
  fb = f(b)
  df = []
  if np.abs(fa) > np.abs(fb):
    c = a
    a = b
    b = c
    c = fa
    fa = fb
    fb = c

  df.append([a, fa])
  df.append([b, fb])

  for i in np.arange(2, nmax+1):
    if np.abs(fa) > np.abs(fb):
      c = a
      a = b
      b = c
      c = fa
      fa = fb
      fb = c

    fc = f((a+b)/2)
    fprime_1 = (fb - fa)/(b-a)
    fprime_2 = (2*fc - fa - fb)/(0.5*(b-a)**2)
    b = a
    fb = fa
    d = fprime_1 / fa
    d = d - 0.5*(fprime_2)/fprime_1
    d = 1/d

    if np.abs(d) < e:
      print("convergence")
      df = pd.DataFrame(df, columns=['x', 'f(x)'])
      return df

    a = a-d
    fa = f(a)
    df.append([a, fa])

  df = pd.DataFrame(df, columns=['x', 'f(x)'])
  return df
```
